```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('download (2).jpeg', cv2.IMREAD_GRAYSCALE)

# Apply GaussianBlur to reduce image noise if necessary
blurred_image = cv2.GaussianBlur(image, (3, 3), 0)

# Sobel Edge Detection
sobelx = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)  # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)  # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 1, ksize=5) # Combined X and Y Sobel Edge Detection

# Display results
plt.subplot(2, 2, 1), plt.imshow(image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(sobelxy, cmap='gray')
plt.title('Sobel X & Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_gabor_filters(image_path):
    # Load the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Error: Unable to load image at {image_path}")
        return

    # Parameters for Gabor filter
    ksize = 31  # Size of the filter
    sigma = 4.0  # Standard deviation of the Gaussian function
    lambd = 10.0  # Wavelength of the sinusoidal component
    gamma = 0.5  # Spatial aspect ratio
    psi = 0  # Phase offset
    angles = [0, np.pi / 4, np.pi / 2, 3 * np.pi / 4]  # Different angles

    # Apply Gabor filters and display results
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    for i, theta in enumerate(angles):
        kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lambd, gamma, psi, ktype=cv2.CV_32F)
        filtered_image = cv2.filter2D(image, cv2.CV_8UC3, kernel)
        plt.subplot(2, 3, i + 2), plt.imshow(filtered_image, cmap='gray')
        plt.title(f'Gabor Filter (Theta={theta:.2f})'), plt.xticks([]), plt.yticks([])

    plt.show()

# Replace 'your_image.jpg' with the path to your image file
apply_gabor_filters('download (2).jpeg')
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = 'download (2).jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded correctly
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Compute the 2D FFT of the image
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)

    # Compute the magnitude spectrum
    magnitude_spectrum = 20 * np.log(np.abs(fshift))

    # Display the original image and its magnitude spectrum
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')  # Hide axis

    plt.subplot(1, 2, 2)
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title('Magnitude Spectrum')
    plt.axis('off')  # Hide axis

    plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def non_max_suppression(image, gradient_direction):
    M, N = image.shape
    Z = np.zeros((M,N), dtype=np.int32)
    angle = gradient_direction * 180. / np.pi
    angle[angle < 0] += 180

    for i in range(1, M-1):
        for j in range(1, N-1):
            try:
                q = 255
                r = 255

                # angle 0
                if (0 <= angle[i,j] < 22.5) or (157.5 <= angle[i,j] <= 180):
                    q = image[i, j+1]
                    r = image[i, j-1]
                # angle 45
                elif (22.5 <= angle[i,j] < 67.5):
                    q = image[i+1, j-1]
                    r = image[i-1, j+1]
                # angle 90
                elif (67.5 <= angle[i,j] < 112.5):
                    q = image[i+1, j]
                    r = image[i-1, j]
                # angle 135
                elif (112.5 <= angle[i,j] < 157.5):
                    q = image[i-1, j-1]
                    r = image[i+1, j+1]

                if (image[i,j] >= q) and (image[i,j] >= r):
                    Z[i,j] = image[i,j]
                else:
                    Z[i,j] = 0

            except IndexError as e:
                pass

    return Z

# Load the image
image_path = 'download (2).jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded correctly
if image is None:
    print(f"Error: Unable to load image at {image_path}")
else:
    # Apply GaussianBlur to reduce image noise if necessary
    blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

    # Compute gradients using Sobel operator
    gradient_x = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)
    gradient_y = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)

    # Calculate magnitude and direction of gradient
    magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
    direction = np.arctan2(gradient_y, gradient_x)

    # Apply non-maximum suppression
    thinned_edges = non_max_suppression(magnitude, direction)

    # Display the original image and the result side by side
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(thinned_edges, cmap='gray')
    plt.title('Thinned Edges')
    plt.axis('off')

    plt.show()
```
